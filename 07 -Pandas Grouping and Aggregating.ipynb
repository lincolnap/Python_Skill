```python
country_respon = df['Country'].value_counts()
country_respon
```

```python
countryusePy = countries['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
countryusePy
```

```python
python_df = pd.concat([country_respon,countryusePy], axis='columns')
python_df
```

```python
#Cambiar el nombre a las colunmas 
python_df.rename(columns={'Country':'Respuestas','LanguageWorkedWith':'Work Python'}, inplace=True)
python_df
```

```python
#para anadir una colunma del porcetaje sin desaparecer los datos de antiguos
python_df['PctKnowsPY'] = (python_df['Work Python']/python_df['Respuestas']) * 100 
python_df
```
